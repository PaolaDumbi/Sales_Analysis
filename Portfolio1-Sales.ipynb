```python
import pandas as pd
import os 
```

### Merging 12 months of sales data into a single file

```python

path = "./Sales_Data"
files = [file for file in os.listdir(path) if not file.startswith('.')] # Ignore hidden files

all_months_data = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+"/"+file)
    all_months_data = pd.concat([all_months_data, current_data])
    
all_months_data.to_csv("all_data.csv", index=False)
    
```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

## CLEAN THE DATA

```python
# DROP rows of NAN
# Find NAN
nan_df = all_data[all_data.isna().any(axis=1)]
display(nan_df.head())

all_data = all_data.dropna(how='all')
all_data.head()
```

```python
# Get rid of text "Or" in the Order Date Column 
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
```

```python
# Make columns correct type
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

## ANALYSIS


Question 1: what was the best month for sales? how much was earned that month? 

```python
##Add a month coulumn
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

```python
## Add a sales column
all_data['Sales']= all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
results=all_data.groupby('Month').sum()
print(results.head())

# results=all_data.groupby('Month').sum()['Sales']#just to have the sales column
```

```python
import matplotlib.pyplot as plt

months= range(1,13)
plt.bar(months,results['Sales'])
plt.xticks(months)
plt.ylabel('ales in USD ($)')
plt.xlabel('Month Number')
plt.show()
```

Question 2: What city has the highest number of sales

```python
# Add a city column
def get_city(adress):
    return adress.split(',')[1]

def get_state(adress):
    return adress.split(',')[2].split(' ')[1]

all_data['City']=all_data['Purchase Address'].apply(lambda x: f"{get_city(x)}({get_state(x)})")

#all_data['City']=all_data['Purchase Address'].apply(lambda x: get_city(x)+ '('+get_state(x)+')')

all_data.head()
```

```python
results= all_data.groupby('City').sum()
results
```

```python
#Ploting
#Results
import matplotlib.pyplot as plt

cities= [city for city, df in all_data.groupby('City')]

plt.bar(cities,results['Sales'])
plt.xticks(cities, rotation='vertical')
plt.ylabel('Sales in USD ($)')
plt.xlabel('City name')
plt.show()
```

Question 4: What products are most often sold together?

```python
# See all the duplicate in column Order ID
df= all_data[all_data['Order ID'].duplicated(keep=False)]#since we want to keep all occurance of the duplicate we pass "False"

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))

df= df[['Order ID','Grouped']].drop_duplicates()

df.head()
```

```python
#count the number of times, the combination of product were bought together
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary

from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

print(row_list)
for key, value in count.most_common(10):
    print(key, value)
```

Question 5: What product sold the most? Why do you think it sold the most?

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

products= [product for product, df in product_group]
plt.bar(products, quantity_ordered)
plt. xticks(products, rotation='vertical', size=10)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.show()
```

```python
# Referenced: https://stackoverflow.com/questions/14762181/adding-a-y-axis-label-to-secondary-y-axis-in-matplotlib

prices= all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

fig.show()
```
